```python
import turtle
def house(a):
    x = 150
    tl = turtle.Turtle()
    turtle.bgcolor('blue')
    tl.hideturtle()
    tl.pensize(3)

    # vẽ nhà
    tl.begin_fill()
    tl.fillcolor('pink')
    tl.right(180)
    tl.forward(x)
    tl.left(90)
    tl.forward(x*1.1)
    tl.left(90)
    tl.forward(x)
    tl.left(90)
    tl.forward(x*1.1)
    tl.left(90)
    tl.end_fill()

    tl.begin_fill()
    tl.fillcolor('orange')
    tl.goto(-x/2,x/2)
    tl.goto(-x,0)
    tl.end_fill()

    tl.penup()
    tl.goto(-x/4,-x*1.1)
    tl.pendown()
    tl.begin_fill()
    tl.fillcolor('grey')
    tl.pendown()
    tl.right(90)
    tl.forward(x*0.6)
    tl.left(90)
    tl.forward(x/2)
    tl.left(90)
    tl.forward(x*0.6)
    tl.end_fill()

    # vẽ cây
    tl.penup()
    tl.goto(x,-x*1.1)
    tl.pendown()
    tl.begin_fill()
    tl.fillcolor('brown')
    tl.right(180)
    tl.forward(x*1.1/3)
    tl.right(90)
    tl.forward(x/6)
    tl.right(90)
    tl.forward(x*1.1/3)
    tl.right(90)
    tl.forward(x/6)
    tl.end_fill()
    tl.penup()
    tl.goto(x*4/3,-x*2.2/3)
    tl.pendown()
    tl.begin_fill()
    tl.fillcolor('green')
    tl.forward(x/2)
    tl.right(120)
    tl.forward(x/2)
    tl.right(120)
    tl.forward(x/2)
    tl.end_fill()
    tl.right(210)
    tl.penup()
    tl.forward(x*(3**0.5)/4)
    tl.pendown()
    tl.begin_fill()
    tl.fillcolor('green')
    tl.left(90)
    tl.forward(x/2)
    tl.right(120)
    tl.forward(x/2)
    tl.right(120)
    tl.forward(x/2)
    tl.end_fill()
    tl.right(210)
    tl.penup()
    tl.forward(x*(3**0.5)/4)
    tl.pendown()
    tl.begin_fill()
    tl.fillcolor('green')
    tl.left(90)
    tl.forward(x/2)
    tl.right(120)
    tl.forward(x/2)
    tl.right(120)
    tl.forward(x/2)
    tl.end_fill()
    tl.right(210)


    # vẽ mặt trời
    tl.penup()
    tl.forward(x*1.5)
    tl.pendown()
    tl.begin_fill()
    tl.fillcolor('yellow')
    tl.circle(x/5)
    tl.end_fill()
    tl.pensize(1)
    tl.right(90)
    tl.forward(x/3)
    tl.backward(x/3)
    tl.left(90)
    tl.circle(x/5,90)
    tl.right(90)
    tl.forward(x/3)
    tl.backward(x/3)
    tl.left(90)
    tl.circle(x/5,90)
    tl.right(90)
    tl.forward(x/3)
    tl.backward(x/3)
    tl.left(90)
    tl.circle(x/5,90)
    tl.right(90)
    tl.forward(x/3)
    tl.backward(x/3)
    tl.left(90)
    tl.circle(x/5,45)
    tl.right(90)
    tl.forward(x/4)
    tl.backward(x/4)
    tl.left(90)
    tl.circle(x/5,90)
    tl.right(90)
    tl.forward(x/4)
    tl.backward(x/4)
    tl.left(90)
    tl.circle(x/5,90)
    tl.right(90)
    tl.forward(x/4)
    tl.backward(x/4)
    tl.left(90)
    tl.circle(x/5,90)
    tl.right(90)
    tl.forward(x/4)
    tl.backward(x/4)
    tl.left(90)

    turtle.mainloop()
    
    
house(100)
```
