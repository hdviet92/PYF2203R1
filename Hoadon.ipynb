```python

from tkinter import *
root = Tk()
root.geometry('550x350')
root.resizable(0, 0)

def tongtien(event):
    
    try:
        slbunbo = int(var_slbunbo.get())
        slbuncha = int(var_slbuncha.get())
        slbundau = int(var_slbundau.get())
        slnuocngot = int(var_slnuocngot.get())

        if slbunbo < 0 or slbuncha < 0 or slbundau < 0 or slnuocngot < 0:
            raise ValueError("ERROR!")
        else:
            bunbo = var_bunbo.get()
            buncha = var_buncha.get()
            bundau = var_bundau.get()
            nuocngot = var_nuocngot.get()
            pttt = var_pttt.get()

            ct = bunbo*slbunbo + buncha*slbuncha + bundau*slbundau + nuocngot*slnuocngot

            label_total['text'] = '{:,}'.format(ct)   

            label_td1['text'] = "Món"
            label_td2['text'] = "Đơn giá"
            label_td3['text'] = "SL"
            label_td4['text'] = "Tiền"

            label_mon['text'] = "Bún bò\n\nBún chả\n\nBún đậu\n\nNước ngọt"

            label_dg['text'] = '{0:,}\n\n{1:,}\n\n{2:,}\n\n{3:,}'.format(bunbo, buncha, bundau, nuocngot)

            label_sl['text'] = '{0}\n\n{1}\n\n{2}\n\n{3}'.format(slbunbo, slbuncha, slbundau, slnuocngot)

            label_dgsl['text'] = '{0:,}\n\n{1:,}\n\n{2:,}\n\n{3:,}'.format(bunbo*slbunbo, buncha*slbuncha,bundau*slbundau, nuocngot*slnuocngot)

            label_vien['text'] = "*******************************************************"

            label_total['text'] = '{:,}'.format(ct)

            if pttt == 'Tiền mặt(-5%)':
                label_muc['text'] = "Tổng tiền\n\nTiền mặt(-5%)\n\nThanh toán"
            elif pttt == 'Chuyển khoản(+5%)':
                label_muc['text'] = "Tổng tiền\n\nChuyển khoản(+5%)\n\nThanh toán"
            else:
                label_muc['text'] = "Thanh toán"

            if pttt == 'Tiền mặt(-5%)':
                label_pttt['text'] = '-{:,.0f}'.format(ct*0.05)
            elif pttt == 'Chuyển khoản(+5%)':
                label_pttt['text'] = '+{:,.0f}'.format(ct*0.05)

            if pttt == 'Tiền mặt(-5%)':
                label_tt['text'] = '{:,.0f}'.format(ct*0.95)
            elif pttt == 'Chuyển khoản(+5%)':
                label_tt['text'] = '{:,.0f}'.format(ct*1.05)

            gender_hoadon['text'] = 'Hóa đơn'
            gender_hoadon['bd'] = 2

            label_tk['text'] = "Cảm ơn Quý khách!!!"

            label_error['text'] = ''
    except:
        label_error['text'] = 'ERROR!'
        label_error['fg'] = 'red'
    
def enter(event):
    event.widget['fg'] = 'red'

def leave(event):
    event.widget['fg'] = 'black'
    
dg_bunbo = 50000
dg_buncha = 40000
dg_bundau = 60000
dg_nuocngot = 10000
var_bunbo = IntVar()
var_buncha = IntVar()
var_bundau = IntVar()
var_nuocngot = IntVar()
gender_option = LabelFrame(root, width=200, heigh=320, text='Tùy chọn', font=('bold'), labelanchor=N)
gender_option.place(x=10, y=0)
label_menu = Label(gender_option, text='MENU', font=('bold'))
label_menu.place(x=70, y=0)
var_slbunbo = Entry(gender_option, justify=RIGHT, width=4)
var_slbunbo.place(x=150, y=30)
var_slbuncha = Entry(gender_option, justify=RIGHT, width=4)
var_slbuncha.place(x=150, y=60)
var_slbundau = Entry(gender_option, justify=RIGHT, width=4)
var_slbundau.place(x=150, y=90)
var_slnuocngot = Entry(gender_option, justify=RIGHT, width=4)
var_slnuocngot.place(x=150, y=120)
bunbo = Checkbutton(gender_option, text='Bún bò', variable=var_bunbo, onvalue=dg_bunbo, offvalue=0, height=2)
bunbo.place(x=10, y=20, anchor=NW)
buncha = Checkbutton(gender_option, text='Bún chả', variable=var_buncha, onvalue=dg_buncha, offvalue=0, height=2)
buncha.place(x=10, y=50, anchor=NW)
bundau = Checkbutton(gender_option, text='Bún đậu', variable=var_bundau, onvalue=dg_bundau, offvalue=0, height=2)
bundau.place(x=10, y=80, anchor=NW)
nuocngot = Checkbutton(gender_option, text='Nước ngọt', variable=var_nuocngot, onvalue=dg_nuocngot, offvalue=0, height=2)
nuocngot.place(x=10, y=110, anchor=NW)

var_pttt = StringVar()
var_pttt.set("0")
label_pttt = Label(gender_option, text='Phương thức thanh toán', font=('bold'))
label_pttt.place(x=10, y=160)
pttt1 = Radiobutton(gender_option, text='Tiền mặt', variable=var_pttt, value='Tiền mặt(-5%)', height=2)
pttt1.place(x=0, y=190, anchor=NW)
pttt2 = Radiobutton(gender_option, text='Chuyển khoản', variable=var_pttt, value='Chuyển khoản(+5%)', height=2)
pttt2.place(x=0, y=220, anchor=NW)

gender_hoadon = LabelFrame(root, width=300, heigh=320, text='', font=('bold'), labelanchor=N, bd=0)
gender_hoadon.place(x=240, y=0)
label_td1 = Label(root, text="", font=('bold'))
label_td1.place(x=255, y=20)
label_td2 = Label(root, text="", font=('bold'))
label_td2.place(x=350, y=20)
label_td3 = Label(root, text="", font=('bold'))
label_td3.place(x=430, y=20)
label_td4 = Label(root, text="", font=('bold'))
label_td4.place(x=490, y=20)
label_mon = Label(root, text="", justify=LEFT)
label_mon.place(x=250, y=50, anchor=NW)
label_dg = Label(root, text="", justify=RIGHT)
label_dg.place(x=400, y=50, anchor=NE)
label_sl = Label(root, text="", justify=RIGHT)
label_sl.place(x=450, y=50, anchor=NE)
label_dgsl = Label(root, text="", justify=RIGHT)
label_dgsl.place(x=530, y=50, anchor=NE)
label_vien = Label(root, text="")
label_vien.place(x=250, y=160)
label_muc = Label(root, text="", justify=LEFT, font=('bold'), fg='red')
label_muc.place(x=250, y=180, anchor=NW)
label_total = Label(root, text="", justify=RIGHT, fg='red')
label_total.place(x=530, y=185, anchor=NE)
label_pttt = Label(root, text="", justify=RIGHT, fg='red')
label_pttt.place(x=530, y=220, anchor=NE)
label_tt = Label(root, text="", justify=RIGHT, fg='red')
label_tt.place(x=530, y=255, anchor=NE)
label_tk = Label(root, text="")
label_tk.place(x=330, y=285)

label_error = Label(root, text="")
label_error.place(x=90, y=270)
button = Button(root, text='Enter', width=4)
button.bind('<Button-1>', tongtien)
button.bind('<Enter>', enter)
button.bind('<Leave>', leave)
button.place(x=90, y=290)

root.mainloop()
 
```
