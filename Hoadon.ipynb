```python
from tkinter import *
root = Tk()

def selected():
    bunbo = checkvar1.get()
    buncha = checkvar2.get()
    bundau = checkvar3.get()
    nuocngot = checkvar4.get()
    if bunbo == 1:
        bunbo = '50,000'
    if buncha == 1:
        buncha = '45,000'
    if bundau == 1:
        bundau = '55,000'
    if nuocngot == 1:
        nuocngot = '10,000'
    label2['text'] = 'Bún bò\t\t\t' + str(bunbo) + '\n\nBún chả\t\t\t' + str(buncha) + '\n\nBún đậu\t\t\t' + str(bundau) + '\n\nNước ngọt\t\t' + str(nuocngot)


def tongtien(event):
    bunbo = checkvar1.get()
    buncha = checkvar2.get()
    bundau = checkvar3.get()
    nuocngot = checkvar4.get()
    thanhtoan = checkvar5.get()
    if bunbo == 1:
        bunbo = 50000
    if buncha == 1:
        buncha = 45000
    if bundau == 1:
        bundau = 55000
    if nuocngot == 1:
        nuocngot = 10000 
    ct = bunbo + buncha + bundau + nuocngot
    message1['text'] = '\nTỔNG TIỀN\t\t{:,}'.format(ct)
    
    if thanhtoan == 'Tiền mặt(-5%)':
        message2['text'] = '\nTiền mặt(-5%)\t\t-{:,.0f}'.format(ct*0.05)
    elif thanhtoan == 'Chuyển khoản(+5%)':
        message2['text'] = '\nC.Khoản(+5%)\t\t+{:,.0f}'.format(ct*0.05)
        
    if thanhtoan == 'Tiền mặt(-5%)':
        message3['text'] = '\nThanh toán\t{:,.0f}'.format(ct*0.95)
    elif thanhtoan == 'Chuyển khoản(+5%)':
        message3['text'] = '\nThanh toán\t{:,.0f}'.format(ct*1.05) 
        
        
def enter(event):
    event.widget['bg'] = 'red'

def leave(event):
    event.widget['bg'] = 'grey'
    
checkvar1 = IntVar()
checkvar2 = IntVar()
checkvar3 = IntVar()
checkvar4 = IntVar()
gender1 = LabelFrame(root, width=450, heigh=800, text='Tùy chòn',font=('bold'))
gender1.grid(row=0,column=0)
label1 = Label(gender1, text='Thực đơn:', font=('bold'))
label1.pack()
bunbo = Checkbutton(gender1, text='Bún bò', variable=checkvar1, onvalue=1, offvalue=0, command=selected, height=2)
bunbo.pack(anchor=W)
buncha = Checkbutton(gender1, text='Bún chả', variable=checkvar2, onvalue=1, offvalue=0, command=selected, height=2)
buncha.pack(anchor=W)
bundau = Checkbutton(gender1, text='Bún đậu', variable=checkvar3, onvalue=1, offvalue=0, command=selected, height=2)
bundau.pack(anchor=W)
nuocngot = Checkbutton(gender1, text='Nước ngọt', variable=checkvar4, onvalue=1, offvalue=0, command=selected, height=2)
nuocngot.pack(anchor=W)


checkvar5 = StringVar()
checkvar5.set("0")
label3 = Label(gender1, text='Thanh toán:', font=('bold'))
label3.pack()
pt1 = Radiobutton(gender1, text='Tiền mặt', variable=checkvar5, value='Tiền mặt(-5%)', height=2)
pt1.pack(anchor=W)
pt2 = Radiobutton(gender1, text='Chuyển khoản', variable=checkvar5, value='Chuyển khoản(+5%)', height=2)
pt2.pack(anchor=W)



gender2 = LabelFrame(root, width=450, heigh=800, text='Hóa đơn',font=('bold'))
gender2.grid(row=0, column=1)
label2 = Label(gender2, text="")
label2.pack(anchor=W)
label4 = Label(gender2, text="\n********************************")
label4.pack()
message1 = Label(gender2, text="")
message1.pack(anchor=W)
message2 = Label(gender2, text="")
message2.pack(anchor=W)
message3 = Label(gender2, text="", font=('bold'))
message3.pack(anchor=W)

button = Button(gender2, text='Enter', width=4, bg='grey')
button.bind('<Button-1>', tongtien)
button.bind('<Enter>', enter)
button.bind('<Leave>', leave)
button.pack()

root.mainloop()
```
