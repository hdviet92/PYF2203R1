```python
class Student:
    def __init__(self, code, fullname, birthday, residence):
        self.code = code
        self.fullname = fullname
        self.birthday = birthday
        self.residence = residence
        
    def info(self):
        print('{0}\t\t{1}\t\t{2}\t\t{3}'.format(self.code, self.fullname, self.birthday, self.residence))
```

```python
class Listsutudent:
    def __init__(self, _list):
        self.list = _list
        
    def initialize(self):
        st1 = Student('220033', 'Huỳnh Đức Việt', '11/09/1992', 'Đà Nẵng')
        st2 = Student('220034', 'Lê Thị Kính', '12/02/1993', 'Đà Nẵng')
        st3 = Student('220035', 'Huỳnh Lê Như Ý', '12/09/2019', 'Đà Nẵng')
        self.list = [st1, st2, st3]
        
    def output(self):
        print("Mã SV\t\tHọ và tên\t\tNgày sinh\t\tQuê quán")
        for i in self.list:
            i.info()
            
    def add(self, std):
        self.list.append(std)
        
    def update(self, code, std):
        '''code: mã sinh viên bị chỉnh sửa
        std: thông tin sinh viên được cập nhật'''
        for i in self.list:
            if i.code == code:
                if std.fullname =='':
                    i.fullname = i.fullname
                else:
                    i.fullname = std.fullname
                if std.birthday =='':
                    i.birthday = i.birthday
                else:
                    i.birthday = std.birthday
                if std.residence =='':
                    i.residence = i.residence
                else:
                    i.residence = std.residence
    
    def delete(self, code):
        for i in self.list:
            if i.code == code:
                self.list.remove(i)
                break

    def search(self, code):
        for i in self.list:
            if i.code == code:
                return i
                
```

```python
class Studentmanagement:
    list_st = None
    def menu(self):
        print("*** QUẢN LÝ DANH SÁCH SINH VIÊN ***")
        print("Nhập 0. Tải danh sách sinh viên")
        print("Nhập 1. Hiển thị danh sách sinh viên")
        print("Nhập 2. Thêm sinh viên             ")
        print("Nhập 3. Cập nhật sinh viên         ")
        print("Nhập 4. Xóa sinh viên              ")
        print("Nhập 5. Tìm kiếm sinh viên         ")
        print("Nhập 6. Thoát                      ")
        print("***********************************")
    
    
    def run(self):
        while True:
            self.menu()
            action = input('Nhập vào lựa chọn: ')
            
            if action == '0':
                self.list_st = Listsutudent(None)
                self.list_st.initialize()
                print('Tải danh sách sinh viên thành công')
                input('Bấm bất kỳ một phím để tiếp tục')
                
            elif action == '1':
                if self.list_st == None:
                    print('Bạn chưa load sinh viên')
                else:
                    self.list_st.output()
                input('Bấm bất kỳ một phím để tiếp tục')
                
            elif action == '2':
                if self.list_st == None:
                    print('Bạn chưa load sinh viên')
                else:                
                    code = input('Nhập mã Sinh viên: ').strip()
                    if self.list_st.search(code) == None:
                        fullname = input('Nhập họ và tên: ').title().strip()
                        birthday = input('Nhập ngày sinh: ').strip()
                        residence = input('Nhập quê quán: ').title().strip()
                        adding_student = Student(code, fullname, birthday, residence)
                        self.list_st.add(adding_student)
                        print('Thêm sinh viên thành công:')
                    else:
                        print('Mã sinh viên này đã có. Vui lòng nhập lại')
                input('Bấm bất kỳ một phím để tiếp tục')
                
            elif action == '3':
                if self.list_st == None:
                    print('Bạn chưa load sinh viên')
                else:
                    code = input('Nhập mã Sinh viên: ').strip()
                    if self.list_st.search(code) != None:
                        fullname = input('Nhập họ và tên: ').title().strip()
                        birthday = input('Nhập ngày sinh: ').strip()
                        residence = input('Nhập quê quán: ').title().strip()
                        updating_student = Student(code, fullname, birthday, residence)
                        self.list_st.update( code, updating_student)
                        print('Cập nhật lại sinh viên thành công:')
                    else:
                        print('Mã sinh viên này đã không có. Vui lòng nhập lại')
                input('Bấm bất kỳ một phím để tiếp tục')
                
            elif action == '4':
                if self.list_st == None:
                    print('Bạn chưa load sinh viên')
                else:
                    code = input('Nhập mã Sinh viên: ').strip()
                    self.list_st.delete(code)
                    print('Xóa sinh viên thành công:')                
                input('Bấm bất kỳ một phím để tiếp tục')
                
            elif action == '5':
                if self.list_st == None:
                    print('Bạn chưa load sinh viên')
                else:
                    code = input('Nhập mã Sinh viên: ').strip()
                    print('Tìm kiếm sinh viên:')
                    find = self.list_st.search(code)
                    if find != None:
                        print("Mã SV\t\tHọ và tên\t\tNgày sinh\t\tQuê quán")
                        find.info()
                    else:
                        print('Không tìm thấy mã sinh viên')
                input('Bấm bất kỳ một phím để tiếp tục')
                
            elif action == '6':
                break
        print('Chương trình kết thúc')
            
```

```python
sm = Studentmanagement()
sm.run()
```
