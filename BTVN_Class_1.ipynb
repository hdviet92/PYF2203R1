```python
# Xây dựng lớp hình tròn có bán kính r. 
# Phương thức: Gồm kiểm tra sự tồn tại hình tròn, tính chu vi và diện tích
import math

class Hinhtron:
    
    # Constructor
    def __init__(self, r):
        self.r = r
        
    def validate(self):
        if self.r > 0:
            return True
        else:
            return False
    
    def area(self):
        if self.validate() is True:
            return math.pow(self.r,2) * math.pi
    
    def lenght(self):
        if self.validate() is True:
            return self.r * 2 * math.pi

r = float(input('Nhập bán kính: '))
ht = Hinhtron(r)
chuvi = ht.lenght()
dientich = ht.area()
print('Chu vi =', chuvi)
print('Diện tích =', dientich)
```

```python
class PTB2:
    
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    def nghiem(self):
        if self.a == 0:
            if self.b == 0:
                if self.c == 0:
                    x = 'vô số nghiệm'
                    x1 = x2 = None
                else:
                    x = 'vô nghiệm'
                    x1 = x2 = None
            else:
                x = 'nghiệm kép'
                x1 = x2 = -self.c/self.b
        else:
            delta = self.b*self.b - 4*self.a*self.c

            if delta < 0:
                x = 'vô nghiệm'
                x1 = x2 = None
            elif delta == 0:
                x = 'nghiệm kép'
                x1 = x2 = -self.b/(2*self.a)
            elif delta > 0:
                x = 'hai nghiệm'
                x1 = (-self.b + delta**0.5)/(2*self.a)
                x2 = (-self.b - delta**0.5)/(2*self.a)
        
        return x, x1, x2
  

    def phuongtrinh(self):
        
        if self.a != 0 and self.b != 0 and self.c != 0:
            pt = (str(self.a)+'x^2' + ' + ' + str(self.b)+'x' +  ' + ' + str(self.c) + ' = 0').replace('1x','x')

        if self.a != 0 and self.b != 0 and self.c == 0:
            pt = (str(self.a)+'x^2' + ' + ' + str(self.b)+'x' + ' = 0').replace('1x','x')
            
        if self.a == 0 and self.b != 0 and self.c != 0:
            pt = (str(self.b)+'x' + ' + ' + str(self.c) + ' = 0').replace('1x','x')

        if self.a != 0 and self.b == 0 and self.c != 0:
            pt = (str(self.a)+'x^2' + ' + ' + str(self.c) + ' = 0').replace('1x','x')

        if self.a == 0 and self.b != 0 and self.c == 0:
            pt = (str(self.b)+'x' + ' = 0').replace('1x','x')

        if self.a != 0 and self.b == 0 and self.c == 0:
            pt = (str(self.a)+'x^2' + ' = 0').replace('1x','x')
        
        if self.a == 0 and self.b == 0:
            pt = (str(self.c) + ' = 0')
        
        pt = pt.replace('+ -','- ')
        
        return pt
            
            
a = int(input('Nhập hệ số 1: '))
b = int(input('Nhập hệ số 2: '))
c = int(input('Nhập hệ số 3: '))
ptb2 = PTB2(a,b,c)
phuongtrinh = ptb2.phuongtrinh()
nghiem = ptb2.nghiem()
print('Phương trình:', phuongtrinh)
print('Nghiệm:', nghiem)
```

```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def distance (self, point):
        '''point là 1 điểm khác, có kiểu dữ liệu là Point'''
        return math.sqrt((self.x - point.x)**2 + (self.y - point.y)**2)
    
    def print(self):
        print("({0}, {1})".format(self.x, self.y))
        
    def info(self):
        return '  ({0}, {1})'.format(self.x, self.y)
    
class Triangle:
    def __init__(self, p1, p2, p3):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        
    def validate(self):
        a = p1.distance(p2)
        b = p1.distance(p3)
        c = p2.distance(p3)
        if a + b > c and a + c > b and c + b > a:
            return True
        else:
            return False
        
    def length(self):
        if self.validate():
            return p1.distance(p2) + p1.distance(p3) + p2.distance(p3)
      
    def area(self):
        if self.validate():
            a = p1.distance(p2)
            b = p1.distance(p3)
            c = p2.distance(p3)
            p = (a+b+c)/2
            return math.sqrt(p*(p-a)*(p-b)*(p-c))

    
p1 = Point(0, 0)
p2 = Point(3, 0)
p3 = Point(3, 4)
tamgiac = Triangle(p1, p2, p3)
print(tamgiac.validate())
print('CV =', tamgiac.length())
print('S =', tamgiac.area())
```

```python
import random
import turtle

n = int(input('Nhập số tọa độ: '))
points = []
rd_diem = [i for i in range (0,400,3)]
colors = ['red', 'orange', 'pink', 'yellow', 'green', 'blue', 'brown']
for i in range(n):
    x = random.choice(rd_diem)
    y = random.choice(rd_diem)
    point = Point(x, y)
    point.print()
    points.append(point)

t= turtle.Turtle()
for p in points:
    t.penup()
    t.goto(p.x, p.y)
    t.pendown()
    t.write(p.info())
    t.begin_fill()
    t.fillcolor(random.choice(colors))
    t.circle(5)
    t.end_fill()
    
turtle.mainloop()
```
