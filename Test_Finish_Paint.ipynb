```python
from tkinter import * 
from tkinter import colorchooser, filedialog, messagebox, ttk
from PIL import Image, ImageTk
import PIL.ImageGrab

eraser_color = 'white'
pen_color = 'black'
pen_width = 1
x1, y1 = 0, 0

def com(event):
    global x1, y1
    x1 = event.x
    y1 = event.y

def line_color(event):
    global x1, y1
    line = canvas.create_line(x1, y1, event.x, event.y, fill=pen_color, width=pen_width, smooth=True, capstyle='round')
    x1 = event.x
    y1 = event.y

def size(event):
    global pen_width
    pen_width = event
    
def color_press(event):
    global pen_color
    pen_color = event
    lb_info.config(bg=event, fg=event)
    
def canvas_color():
    global eraser_color
    global pen_color
    a = colorchooser.askcolor()
    if a:
        canvas.config(bg=a[1])
        pen_color = eraser_color = a[1]
        
def eraser():
    global pen_color
    pen_color = eraser_color
    lb_info.config(bg='white', text="Eraser", fg='green')

def clear():
    global pen_color
    global eraser_color
    canvas.delete('all')
    canvas.config(bg='white')
    eraser_color = 'white'
    pen_color = 'black'
    lb_info.config(bg='black', text="", fg='black')
    
def save():
    try:
        filename = filedialog.asksaveasfilename(defaultextension='.jpg', filetypes=[("JPG", "*.jpg")])
        x = canvas.winfo_rootx()
        y = canvas.winfo_rooty()
        w = x + canvas.winfo_width()
        h = y + canvas.winfo_height()
        ImageGrab.grab(bbox=(x, y, w, h)).save(filename)
    except:
        messagebox.showerror("Error", message="Some error occurs !")
    
def open():
    try:  
        file_name = filedialog.askopenfilename()
        if not file_name: 
            return 
        img = Image.open(file_name)
        canvas.image = ImageTk.PhotoImage(img)
        iden = canvas.create_image(0, 0, image = canvas.image, anchor="nw", tag="obj")
    except: 
        messagebox.showerror("Error", "Can't open this file")     
        
def enter(event):
    event.widget['fg'] = 'red'
    
def leave(event):
    event.widget['fg'] = 'black'
    
root = Tk()
root.resizable(False, False)
root.title('Paint')
root.config(bg='white')

lbframe_pencil = LabelFrame(root, labelanchor='n', text="Pencil", bd=4, relief=RIDGE, bg='white')
lbframe_pencil.grid(row=0, column=0, pady=5, padx=10)
lb_info = Label(lbframe_pencil, width=8, bg='black')
lb_info.grid(row=0, column=0)

lbframe_colors = LabelFrame(root, labelanchor='n', text="Color", bd=4, relief=RIDGE, bg='white')
lbframe_colors.grid(row=1, column=0, pady=5, padx=10)
colors = ['black', 'gray', 'red', 'green', 'blue', 'purple', 'pink', 
          'lightblue', 'orange', 'cyan', 'magenta', 'yellow','violet','white']
r = c = 0
for color in colors:
    Button(lbframe_colors, bg=color, width=3, borderwidth=2, relief=RIDGE, command=lambda col=color: color_press(col)).grid(row=r, column=c)
    r += 1
    if r == 7:
        r = 0
        c = 1
btn_eraser = Button(lbframe_colors, text="Eraser", width=7, bd=4, relief=RIDGE, command=eraser)
btn_eraser.grid(row=7, column=0, columnspan=2, pady=5)        
btn_eraser.bind('<Enter>', enter)
btn_eraser.bind('<Leave>', leave)

frame_func = Frame(root)
frame_func.grid(row=2, column=0, pady=5, padx=10)
btn_clear = Button(frame_func, text="Clear", width=8, bd=4, relief=RIDGE, command=clear)
btn_clear.grid(row=0, column=0)
btn_clear.bind('<Enter>', enter)
btn_clear.bind('<Leave>', leave)
btn_save = Button(frame_func, text="Save", width=8, bd=4, relief=RIDGE, command=save)
btn_save.grid(row=1, column=0)
btn_save.bind('<Enter>', enter)
btn_save.bind('<Leave>', leave)
btn_canvas = Button(frame_func, text="Canvas", width=8, bd=4, relief=RIDGE, command=canvas_color)
btn_canvas.grid(row=2, column=0)
btn_canvas.bind('<Enter>', enter)
btn_canvas.bind('<Leave>', leave)
btn_open = Button(frame_func, text="Open", width=8, bd=4, relief=RIDGE, command=open)
btn_open.grid(row=3, column=0)
btn_open.bind('<Enter>', enter)
btn_open.bind('<Leave>', leave)

lbframe_size = LabelFrame(root, labelanchor='n', text="Size", relief=RIDGE, bd=4, bg='white')
lbframe_size.grid(row=3, column=0, pady=5, padx=10)
scale_size = ttk.Scale(lbframe_size, orient=VERTICAL, from_=70, to=1, length=170, command=size)
scale_size.grid(row=0, column=0, padx=15)

canvas = Canvas(root, bd=4, relief=GROOVE, bg='white', height=650, width=900, cursor='dot')
canvas.grid(row=0, column=1, rowspan=4, pady=5)
canvas.bind('<B1-Motion>', line_color)
canvas.bind('<Button-1>', com)

root.mainloop()
```
