```python
import tkinter as tk
from tkinter import ttk

root = tk.Tk()

# config the root window
root.geometry('300x200')
root.resizable(False, False)
root.title('Cộng trừ nhân chia hai số')

lbNumber1 = tk.Label(root, text= "Số thứ nhất", font=("Helvetica", 12))
lbNumber1.place(x=10, y=10)

def validate1(event):
    try:
        text_input = int(event.widget.get())
        lbError1['text'] = ""
    except:
        lbError1['text'] = "Số thứ nhất chưa đúng"
        
def validate2(event):
    try:
        text_input = int(event.widget.get())
        lbError2['text'] = ""
    except:
        lbError2['text'] = "Số thứ hai chưa đúng"
    
txtNumber1 = tk.Entry(root,  font=("Helvetica", 12))
txtNumber1.place(x=100, y=10)
txtNumber1.bind("<FocusOut>", validate1)

lbError1 = tk.Label(root, font=("Helvetica", 8, 'italic'), fg='Red')
lbError1.place(x=100, y=30)

lbNumber2 = tk.Label(root, text= "Số thứ hai", font=("Helvetica", 12))
lbNumber2.place(x=10, y=60)

txtNumber2 = tk.Entry(root,  font=("Helvetica", 12))
txtNumber2.place(x=100, y=60)
txtNumber2.bind("<FocusOut>", validate2)

lbError2 = tk.Label(root, font=("Helvetica", 8, 'italic'), fg='Red')
lbError2.place(x=100, y=80)

lbAction = tk.Label(root, text= "Thao tác", font=("Helvetica", 12))
lbAction.place(x=10, y=110)

selected_action = tk.StringVar()
selected_action.set("Cộng")
cbAction = ttk.Combobox(root, textvariable=selected_action)
cbAction.place(x=100, y=110)
cbAction['values'] = ["Cộng", "Trừ", "Nhân", "Chia"]
cbAction['state'] = 'readonly'

def calculation(event):
    try:
        btCalculate.focus()
        if(txtNumber1.get()==""):
            lbError1['text'] = "Số thứ nhất chưa đúng"

        if(txtNumber2.get()==""):
            lbError2['text'] = "Số thứ hai chưa đúng"

        if lbError2['text']!="" or lbError1['text']!="":
            lbResult['text'] = ""
            return
        action = selected_action.get()
        lbResult['fg'] = 'green'
        if action == "Cộng":        
            lbResult['text'] = int(txtNumber1.get()) + int(txtNumber2.get())
        elif action == "Trừ":
            lbResult['text'] = int(txtNumber1.get()) - int(txtNumber2.get())
        elif action == "Nhân":
            lbResult['text'] = int(txtNumber1.get()) * int(txtNumber2.get())
        else:
            if int(txtNumber2.get())!=0:
                lbResult['text'] = int(txtNumber1.get()) / int(txtNumber2.get())
            else:
                lbResult['text'] = "NAN"
                lbResult['fg'] = 'red'
    except:
        lbResult['text'] = ""
            
btCalculate = tk.Button(root, text ="Tính", font=("Helvetica", 12))
btCalculate.place(x=100, y=140)
btCalculate.bind("<Button-1>", calculation)

lbResult = tk.Label(root, text ="", font=("Helvetica", 12))
lbResult.place(x=160, y=145)

root.mainloop()
```
