```python
from tkinter import *

text_display = ''
first_number = 0
second_number = 0
action = ''

def number_press(event):
    global text_display
    pressed = event.widget['text']
    if text_display=='' or text_display=='0':
        if pressed == '0':
            text_display='0'
        else:
            if text_display=='0':
                text_display=''
            text_display += pressed    
    else:
        if text_display=='0':
            text_display=''
        text_display += pressed
    screen.delete(0,'end')
    screen.insert(0, text_display)

def action_press(event):
    global first_number 
    first_number = int(screen.get())
    global action
    action = event.widget['text']
    global text_display
    text_display = ''

def finish_press(event):
    global second_number
    second_number = int(screen.get())
    if action=='+':
        result = first_number+second_number
    elif action=='-':
        result = first_number-second_number    
    elif action=='*':
        result = first_number*second_number
    else:
        if second_number!=0:
            result = first_number/second_number
        else:
            result= 'NaN'
    screen.delete(0,'end')
    screen.insert(0, result)
    
def clear_press(event):
    global text_display
    text_display = ''
    global first_number
    first_number = 0
    global second_number
    second_number = 0
    global action
    action = ''
    screen.delete(0,'end')
    screen.insert(0, '0')
    
cal = Tk()
cal.title('Calculator')
cal.geometry('305x390')
cal.resizable(0, 0)

screen = Entry(cal, font=('Helvetica',17, 'bold'), justify='right', bg='cyan')
screen.grid(row=0,column=0, columnspan=5, padx=5, pady=5, ipadx=10,ipady=10)
screen.insert(0, '0')

keyboards = {'MC': 0, 'MR': 0, 'MS': 0, 'M+': 0, 'M-': 0, '←': 0, 'CE': 0, 'C': 4, '±': 0, '√': 0, '7': 1, '8': 1,
            '9': 1, '/': 2, '%': 0, '4': 1, '5': 1, '6': 1, '*': 2, '1/x': 0, '1': 1, '2': 1, '3': 1, '-': 2, '=': 3,
            '0': 1, '': 0, '.': 0, '+': 2}

keys = list(keyboards.keys())
row = 1
column = 0
for i in range (len(keys)):
    if i%5==0 and i!=0:
        row +=1
    if i==26:
        continue
    if i == 25:
        btn = Button(cal, text=keys[i], font=('Helvetica', 10, 'bold'), heigh=2, width=13)
        btn.grid(row=row, column=i%5, padx=5, pady=5, columnspan=2)
    elif i == 24:
        btn = Button(cal, text=keys[i], font=('Helvetica', 10, 'bold'), heigh=5, width=5)
        btn.grid(row=row, column=i%5, padx=5, pady=5, rowspan=2, ipady=3)
    else:
        btn = Button(cal, text=keys[i], font=('Helvetica', 10, 'bold'), heigh=2, width=5)
        btn.grid(row=row, column=i%5, padx=5, pady=5)
    if keyboards[keys[i]] == 1:
        btn.bind('<Button-1>', number_press)
    elif keyboards[keys[i]] == 2:
        btn.bind('<Button-1>',action_press)
    elif keyboards[keys[i]] == 3:
        btn.bind('<Button-1>',finish_press)
    elif keyboards[keys[i]] == 4:
        btn.bind('<Button-1>',clear_press)
              
cal.mainloop()
```
