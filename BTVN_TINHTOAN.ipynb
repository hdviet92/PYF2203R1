```python
from tkinter import *

window = Tk()
window.title('BTVN')
window.geometry('400x250') 

def cong(event):
    try:
        number1 = int(txtnumber1.get())
        number2 = int(txtnumber1.get())
        message['text'] = int(txtnumber1.get()) + int(txtnumber2.get())
    except:
        message['text'] = 'Lỗi nhập dữ liệu'
        
def tru(event):
    try:
        number1 = int(txtnumber1.get())
        number2 = int(txtnumber1.get())
        message['text'] = int(txtnumber1.get()) - int(txtnumber2.get())
    except:
        message['text'] = 'Lỗi nhập dữ liệu'

def nhan(event):
    try:
        number1 = int(txtnumber1.get())
        number2 = int(txtnumber1.get())
        message['text'] = int(txtnumber1.get()) * int(txtnumber2.get())
    except:
        message['text'] = 'Lỗi nhập dữ liệu'

def chia(event):
    try:
        number1 = int(txtnumber1.get())
        number2 = int(txtnumber1.get())
        message['text'] = int(txtnumber1.get()) / int(txtnumber2.get())
    except:
        message['text'] = 'Lỗi nhập dữ liệu'

def enter(event):
    event.widget['bg'] = 'red'
    
def leave(event):
    event.widget['bg'] = 'grey'
    
message = Label(window, text='Output:',bg= 'powder blue', width=56, height=3, relief=RIDGE)
message.grid(column=0, row=0, columnspan=5)

clear = Label(window, width=56, height=1).grid(column=0, row=1, columnspan=5,)    
txtnumber1 = Entry(window, width = 20, justify='right')
txtnumber2 = Entry(window,width = 20, justify='right')
txtnumber1.grid(column=0, row=2,columnspan=2)
txtnumber2.grid(column=3, row=2,columnspan=2)
clear = Label(window, width=56, height=1).grid(column=0, row=3, columnspan=5)  

button1 = Button(window, text='+', width = 4, bg = 'grey')
button1.bind('<Button-1>', cong)
button1.bind('<Enter>', enter)
button1.bind('<Leave>', leave)
button1.grid(column=0, row=4)

button2 = Button(window, text='-', width = 4, bg = 'grey')
button2.bind('<Button-1>', tru)
button2.bind('<Enter>', enter)
button2.bind('<Leave>', leave)
button2.grid(column=1, row=4)

button3 = Button(window, text='*', width = 4, bg = 'grey')
button3.bind('<Button-1>', nhan)
button1.bind('<Enter>', enter)
button1.bind('<Leave>', leave)
button3.grid(column=3, row=4)

button4 = Button(window, text='/', width = 4, bg = 'grey')
button4.bind('<Button-1>', chia)
button1.bind('<Enter>', enter)
button1.bind('<Leave>', leave)
button4.grid(column=4, row=4)

window.mainloop()
```
