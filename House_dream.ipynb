```python
import turtle
tl = turtle.Turtle()
turtle.bgcolor('blue')
tl.hideturtle()
tl.pensize(3)

# vẽ nhà
tl.begin_fill()
tl.fillcolor('pink')
tl.right(180)
tl.forward(150)
tl.left(90)
tl.forward(165)
tl.left(90)
tl.forward(150)
tl.left(90)
tl.forward(165)
tl.left(90)
tl.end_fill()

tl.begin_fill()
tl.fillcolor('orange')
tl.goto(-75,75)
tl.goto(-150,0)
tl.end_fill()

tl.penup()
tl.goto(-37.5,-165)
tl.pendown()
tl.begin_fill()
tl.fillcolor('grey')
tl.pendown()
tl.right(90)
tl.forward(90)
tl.left(90)
tl.forward(75)
tl.left(90)
tl.forward(90)
tl.end_fill()

# vẽ cây
tl.penup()
tl.goto(150,-165)
tl.pendown()
tl.begin_fill()
tl.fillcolor('brown')
tl.right(180)
tl.forward(55)
tl.right(90)
tl.forward(25)
tl.right(90)
tl.forward(55)
tl.right(90)
tl.forward(25)
tl.end_fill()
tl.penup()
tl.goto(200,-110)
tl.pendown()
tl.begin_fill()
tl.fillcolor('green')
tl.forward(75)
tl.right(120)
tl.forward(75)
tl.right(120)
tl.forward(75)
tl.end_fill()
tl.right(210)
tl.penup()
tl.forward(37.5*(3**0.5))
tl.pendown()
tl.begin_fill()
tl.fillcolor('green')
tl.left(90)
tl.forward(75)
tl.right(120)
tl.forward(75)
tl.right(120)
tl.forward(75)
tl.end_fill()
tl.right(210)
tl.penup()
tl.forward(37.5*(3**0.5))
tl.pendown()
tl.begin_fill()
tl.fillcolor('green')
tl.left(90)
tl.forward(75)
tl.right(120)
tl.forward(75)
tl.right(120)
tl.forward(75)
tl.end_fill()
tl.right(210)


# vẽ mặt trời
tl.penup()
tl.forward(225)
tl.pendown()
tl.begin_fill()
tl.fillcolor('yellow')
tl.circle(30)
tl.end_fill()
tl.pensize(1)
tl.right(90)
tl.forward(50)
tl.backward(50)
tl.left(90)
tl.circle(30,90)
tl.right(90)
tl.forward(50)
tl.backward(50)
tl.left(90)
tl.circle(30,90)
tl.right(90)
tl.forward(50)
tl.backward(50)
tl.left(90)
tl.circle(30,90)
tl.right(90)
tl.forward(50)
tl.backward(50)
tl.left(90)
tl.circle(30,45)
tl.right(90)
tl.forward(37.5)
tl.backward(37.5)
tl.left(90)
tl.circle(30,90)
tl.right(90)
tl.forward(37.5)
tl.backward(37.5)
tl.left(90)
tl.circle(30,90)
tl.right(90)
tl.forward(37.5)
tl.backward(37.5)
tl.left(90)
tl.circle(30,90)
tl.right(90)
tl.forward(37.5)
tl.backward(37.5)
tl.left(90)

turtle.mainloop()
```
